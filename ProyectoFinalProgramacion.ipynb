```python
# üëâ APP FLASK
app = Flask(__name__)

@app.route("/", methods=["GET", "POST"])
def home():
    resultado = ""
    probs = [0, 0, 0]
    if request.method == "POST" and modelo_cargado:
        local = int(request.form["local"])
        visitante = int(request.form["visitante"])
        pred_probs = modelo.predict_proba([[local, visitante]])[0]
        pred = np.random.choice([0, 1, 2], p=pred_probs)
        resultados = {0: "Empate", 1: "Gana Local", 2: "Gana Visitante"}
        resultado = resultados[pred]
        probs = [round(p * 100, 2) for p in pred_probs]
    return render_template_string(html_template, resultado=resultado, probs=probs)
```

```python
# üëâ INICIAR SERVIDOR + NGROK
def run():
    app.run()

public_url = ngrok.connect(5000)
print(f"‚úÖ Abre esta URL para usar la app: {public_url}")

thread = Thread(target=run)
thread.start()
```
